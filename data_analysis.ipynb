```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform
import numpy as np
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

data = pd.read_csv('C:\\Users\\ksk03\\PycharmProjects\\Web\\data_df.csv',sep=',', thousands=',')
#data_ = data.astype({'오늘 확진자':int,'총 확진자': int, '총 사망자':int})
avg = data.groupby('지역')['오늘 확진자'].mean()
print(avg[avg>avg.mean()])
```

```python
data_graph = data.groupby('지역')['오늘 확진자','총 사망자'].mean().astype({'오늘 확진자':int, '총 사망자':int})
ax = data_graph.plot(kind='bar', title='코로나 발생 추이', figsize=(12, 4), legend=True, fontsize=12)
ax.set_xlabel('지역', fontsize=12)
ax.set_ylabel('인구 수', fontsize=12)
ax.legend(['오늘 확진자','총 사망자'], fontsize=12)
```

```python
data_total_graph = data.groupby('지역')['총 확진자'].mean().astype({'총 확진자':int})
g2 = data_total_graph.plot(kind='bar', title='지역별 코로나 확진자수 차이', figsize=(12, 4), fontsize=12)
g2.set_xlabel('지역', fontsize=12)          
g2.set_ylabel('인구 수', fontsize=12)
```
